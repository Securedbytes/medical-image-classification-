```python
"""
=============================================================================
CNN MEDICAL IMAGE CLASSIFICATION - ASSIGNMENT VERSION
=============================================================================
Optimized for academic assignments:
- Fast training (20-60 minutes)
- All required metrics and visualizations
- Simplified for clarity
- Perfect for reports and presentations
=============================================================================
"""

# =============================================================================
# 1. IMPORTS AND SETUP
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json
import warnings
import time
from datetime import datetime
from PIL import Image  # FIX: Added for image loading
warnings.filterwarnings('ignore')

# Deep Learning
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, CSVLogger

# Machine Learning Metrics
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_curve, auc
)
from sklearn.preprocessing import label_binarize

# Set random seed for reproducibility
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# Track start time
start_time_total = time.time()

print("="*80)
print("CNN MEDICAL IMAGE CLASSIFICATION - ASSIGNMENT VERSION")
print("="*80)
print(f"Start Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"TensorFlow Version: {tf.__version__}")
print(f"Random Seed: {RANDOM_SEED} (for reproducibility)")
print("="*80)

# =============================================================================
# 2. CONFIGURATION (Assignment-Optimized)
# =============================================================================

class Config:
    """Configuration optimized for fast training"""
    
    # File paths
    TRAIN_CSV = 'labels_train.csv'
    VAL_CSV = 'labels_val.csv'
    TEST_CSV = 'manifest_public.csv'
    
    TRAIN_IMAGE_DIR = 'images_train/'
    VAL_IMAGE_DIR = 'images_val/'
    TEST_IMAGE_DIR = 'images/'
    
    RESULTS_DIR = './results_cnn_assignment/'
    MODEL_SAVE_PATH = 'cnn_model.h5'
    
    # Dataset parameters
    NUM_CLASSES = 11
    
    # Image parameters (small = fast)
    IMG_HEIGHT = 32
    IMG_WIDTH = 32
    IMG_CHANNELS = 3
    
    # Training parameters (fast training)
    BATCH_SIZE = 256          # Large batch = fewer iterations
    EPOCHS = 15               # Enough to see convergence
    LEARNING_RATE = 0.001
    
    # Use subset for speed (10% of data)
    USE_SUBSET = True         # Set False for full dataset
    SUBSET_FRACTION = 0.10    # 10% = ~3,500 images
    
    # Data augmentation (light)
    ROTATION_RANGE = 10
    WIDTH_SHIFT_RANGE = 0.1
    HEIGHT_SHIFT_RANGE = 0.1
    ZOOM_RANGE = 0.1
    HORIZONTAL_FLIP = True
    
    CLASS_NAMES = {
        0: 'Class 0', 1: 'Class 1', 2: 'Class 2', 3: 'Class 3',
        4: 'Class 4', 5: 'Class 5', 6: 'Class 6', 7: 'Class 7',
        8: 'Class 8', 9: 'Class 9', 10: 'Class 10'
    }

config = Config()
os.makedirs(config.RESULTS_DIR, exist_ok=True)

print(f"\n  Configuration:")
print(f"   Image Size: {config.IMG_WIDTH}x{config.IMG_HEIGHT}")
print(f"   Batch Size: {config.BATCH_SIZE}")
print(f"   Epochs: {config.EPOCHS}")
print(f"   Using Subset: {config.USE_SUBSET} ({config.SUBSET_FRACTION*100}%)")
print(f"   Random Seed: {RANDOM_SEED}")

# =============================================================================
# 3. LOAD DATASET
# =============================================================================

print("\n" + "="*80)
print("SECTION 3: DATASET HANDLING")
print("="*80)

# Load CSV files
train_df = pd.read_csv(config.TRAIN_CSV)
val_df = pd.read_csv(config.VAL_CSV)
test_df = pd.read_csv(config.TEST_CSV)

print(f"\n Original Dataset Size:")
print(f"   Training:   {len(train_df):,} samples")
print(f"   Validation: {len(val_df):,} samples")
print(f"   Test:       {len(test_df):,} samples")

# Apply subset for fast training
if config.USE_SUBSET:
    print(f"\n Speed Mode: Using {config.SUBSET_FRACTION*100}% of data")
    train_df = train_df.sample(frac=config.SUBSET_FRACTION, random_state=RANDOM_SEED)
    val_df = val_df.sample(frac=config.SUBSET_FRACTION, random_state=RANDOM_SEED)
    
    print(f"   Training:   {len(train_df):,} samples (subset)")
    print(f"   Validation: {len(val_df):,} samples (subset)")

# Convert labels to strings for generator
train_df['label'] = train_df['label'].astype(str)
val_df['label'] = val_df['label'].astype(str)

# Class distribution analysis
print("\n Class Distribution:")
train_class_counts = train_df['label'].value_counts().sort_index()
for cls in sorted(train_class_counts.index):
    count = train_class_counts[cls]
    percentage = (count / len(train_df)) * 100
    print(f"   Class {cls}: {count:4,} samples ({percentage:5.2f}%)")

# Calculate imbalance
max_count = train_class_counts.max()
min_count = train_class_counts.min()
imbalance_ratio = max_count / min_count
print(f"\n  Class Imbalance Ratio: {imbalance_ratio:.2f}:1")

# Visualize class distribution
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
train_class_counts.plot(kind='bar', color='steelblue', alpha=0.8, edgecolor='black', ax=ax)
ax.set_xlabel('Class Label', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Samples', fontsize=12, fontweight='bold')
ax.set_title('Training Set - Class Distribution', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(f'{config.RESULTS_DIR}class_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n Dataset handling complete!")
print(f"   - Train/Val/Test Split: {len(train_df)}/{len(val_df)}/{len(test_df)}")
print(f"   - Split Type: Predefined (separate CSV files)")
print(f"   - Fixed Random Seed: {RANDOM_SEED}")

# =============================================================================
# 4. DATA PREPROCESSING & AUGMENTATION
# =============================================================================

print("\n" + "="*80)
print("SECTION 4: DATA PREPROCESSING")
print("="*80)

# Data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=config.ROTATION_RANGE,
    width_shift_range=config.WIDTH_SHIFT_RANGE,
    height_shift_range=config.HEIGHT_SHIFT_RANGE,
    zoom_range=config.ZOOM_RANGE,
    horizontal_flip=config.HORIZONTAL_FLIP,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

print(f" Data Preprocessing:")
print(f"   Input: Raw images from {config.TRAIN_IMAGE_DIR}")
print(f"   Output: Normalized {config.IMG_WIDTH}x{config.IMG_HEIGHT} tensors")
print(f"   Normalization: Pixel values scaled to [0, 1]")
print(f"\n   Augmentation Parameters:")
print(f"   - Rotation: ±{config.ROTATION_RANGE}°")
print(f"   - Width Shift: ±{config.WIDTH_SHIFT_RANGE*100}%")
print(f"   - Height Shift: ±{config.HEIGHT_SHIFT_RANGE*100}%")
print(f"   - Zoom: ±{config.ZOOM_RANGE*100}%")
print(f"   - Horizontal Flip: {config.HORIZONTAL_FLIP}")

# Create generators
color_mode = 'rgb' if config.IMG_CHANNELS == 3 else 'grayscale'

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=config.TRAIN_IMAGE_DIR,
    x_col='file',
    y_col='label',
    target_size=(config.IMG_HEIGHT, config.IMG_WIDTH),
    batch_size=config.BATCH_SIZE,
    class_mode='categorical',
    color_mode=color_mode,
    shuffle=True,
    seed=RANDOM_SEED
)

val_generator = val_datagen.flow_from_dataframe(
    dataframe=val_df,
    directory=config.VAL_IMAGE_DIR,
    x_col='file',
    y_col='label',
    target_size=(config.IMG_HEIGHT, config.IMG_WIDTH),
    batch_size=config.BATCH_SIZE,
    class_mode='categorical',
    color_mode=color_mode,
    shuffle=False
)

print(f"\n Data generators created:")
print(f"   Training batches: {len(train_generator)}")
print(f"   Validation batches: {len(val_generator)}")

# =============================================================================
# 5. BUILD CNN MODEL
# =============================================================================

print("\n" + "="*80)
print("SECTION 5: CNN MODEL ARCHITECTURE")
print("="*80)

def build_simple_cnn(input_shape, num_classes):
    """
    Simple CNN for assignment demonstration
    Architecture: 3 Conv blocks + 2 Dense layers
    """
    model = models.Sequential([
        # Block 1
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', 
                     input_shape=input_shape, name='conv1'),
        layers.MaxPooling2D((2, 2), name='pool1'),
        layers.Dropout(0.25, name='dropout1'),
        
        # Block 2
        layers.Conv2D(64, (3, 3), activation='relu', padding='same', name='conv2'),
        layers.MaxPooling2D((2, 2), name='pool2'),
        layers.Dropout(0.25, name='dropout2'),
        
        # Block 3
        layers.Conv2D(128, (3, 3), activation='relu', padding='same', name='conv3'),
        layers.MaxPooling2D((2, 2), name='pool3'),
        layers.Dropout(0.25, name='dropout3'),
        
        # Dense layers
        layers.Flatten(name='flatten'),
        layers.Dense(128, activation='relu', name='dense1'),
        layers.Dropout(0.5, name='dropout4'),
        layers.Dense(num_classes, activation='softmax', name='output')
    ])
    
    return model

# Build model
input_shape = (config.IMG_HEIGHT, config.IMG_WIDTH, config.IMG_CHANNELS)
model = build_simple_cnn(input_shape, config.NUM_CLASSES)

# Compile
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=config.LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n  CNN Architecture:")
print(f"   Input Shape: {input_shape}")
print(f"   Architecture: 3 Conv Blocks + 2 Dense Layers")
print(f"   Activation: ReLU (hidden layers), Softmax (output)")
print(f"   Optimizer: Adam (learning_rate={config.LEARNING_RATE})")
print(f"   Loss Function: Categorical Crossentropy")

model.summary()

# Count parameters
total_params = model.count_params()
print(f"\n Model Parameters: {total_params:,}")

# =============================================================================
# 6. TRAIN MODEL
# =============================================================================

print("\n" + "="*80)
print("SECTION 6: MODEL TRAINING")
print("="*80)

# Callbacks
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    ),
    ModelCheckpoint(
        filepath=os.path.join(config.RESULTS_DIR, config.MODEL_SAVE_PATH),
        monitor='val_accuracy',
        save_best_only=True,
        verbose=0
    ),
    CSVLogger(
        os.path.join(config.RESULTS_DIR, 'training_log.csv'),
        append=False
    )
]

print(f"  Training Configuration:")
print(f"   Epochs: {config.EPOCHS}")
print(f"   Batch Size: {config.BATCH_SIZE}")
print(f"   Steps per Epoch: {len(train_generator)}")
print(f"   Validation Steps: {len(val_generator)}")
print(f"   Early Stopping: Patience=5")
print("="*80)

# Train
train_start = time.time()

history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=config.EPOCHS,
    validation_data=val_generator,
    validation_steps=len(val_generator),
    callbacks=callbacks,
    verbose=1
)

train_time = time.time() - train_start

print("\n Training Completed!")
print(f"   Training Time: {train_time:.2f}s ({train_time/60:.2f} min)")
print(f"   Epochs Trained: {len(history.history['loss'])}")
print(f"   Best Epoch: {np.argmax(history.history['val_accuracy']) + 1}")

# =============================================================================
# 7. PLOT TRAINING HISTORY
# =============================================================================

print("\n" + "="*80)
print("SECTION 7: LEARNING CURVES")
print("="*80)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Accuracy
axes[0].plot(history.history['accuracy'], 'b-', linewidth=2, label='Training', marker='o')
axes[0].plot(history.history['val_accuracy'], 'r-', linewidth=2, label='Validation', marker='s')
axes[0].set_title('Model Accuracy vs Epochs', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Accuracy', fontsize=12)
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# Loss
axes[1].plot(history.history['loss'], 'b-', linewidth=2, label='Training', marker='o')
axes[1].plot(history.history['val_loss'], 'r-', linewidth=2, label='Validation', marker='s')
axes[1].set_title('Model Loss vs Epochs', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Loss', fontsize=12)
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f'{config.RESULTS_DIR}training_curves.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Learning curves saved")

# =============================================================================
# 8. EVALUATE MODEL
# =============================================================================

print("\n" + "="*80)
print("SECTION 8: PERFORMANCE EVALUATION")
print("="*80)

# Load best model
model = keras.models.load_model(os.path.join(config.RESULTS_DIR, config.MODEL_SAVE_PATH))

# Predictions
print("\n Generating predictions...")
y_pred_proba = model.predict(val_generator, steps=len(val_generator), verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)
y_true = val_generator.classes

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)

print("\n" + "="*80)
print("PERFORMANCE METRICS")
print("="*80)
print(f"\n Overall Performance:")
print(f"   Accuracy  = {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"   Precision = {precision:.4f} ({precision*100:.2f}%)")
print(f"   Recall    = {recall:.4f} ({recall*100:.2f}%)")
print(f"   F1-Score  = {f1:.4f} ({f1*100:.2f}%)")

print(f"\n Metric Definitions:")
print(f"   Accuracy  = (TP + TN) / Total")
print(f"   Precision = TP / (TP + FP)")
print(f"   Recall    = TP / (TP + FN)")
print(f"   F1-Score  = 2 × (Precision × Recall) / (Precision + Recall)")

# =============================================================================
# PERFORMANCE SUMMARY TABLE
# =============================================================================

# Get final training metrics from history
train_accuracy = history.history['accuracy'][-1]
train_loss = history.history['loss'][-1]
val_accuracy = history.history['val_accuracy'][-1]
val_loss = history.history['val_loss'][-1]

# Calculate AUC
y_true_bin_temp = label_binarize(y_true, classes=range(config.NUM_CLASSES))
fpr_temp, tpr_temp = dict(), dict()
fpr_temp["micro"], tpr_temp["micro"], _ = roc_curve(y_true_bin_temp.ravel(), y_pred_proba.ravel())
roc_auc_micro = auc(fpr_temp["micro"], tpr_temp["micro"])

# Create performance summary table
print("\n" + "="*80)
print("PERFORMANCE SUMMARY TABLE")
print("="*80)

# Create DataFrame
performance_data = {
    'Metric': ['Accuracy', 'Loss', 'Precision', 'Recall', 'F1-Score', 'AUC'],
    'Training': [
        f'{train_accuracy*100:.2f}%',
        f'{train_loss:.4f}',
        '-',
        '-',
        '-',
        '-'
    ],
    'Validation': [
        f'{val_accuracy*100:.2f}%',
        f'{val_loss:.4f}',
        f'{precision*100:.2f}%',
        f'{recall*100:.2f}%',
        f'{f1*100:.2f}%',
        f'{roc_auc_micro:.4f}'
    ],
    'Definition': [
        'Correct predictions / Total',
        'Cross-entropy divergence',
        'TP / (TP+FP) weighted',
        'TP / (TP+FN) weighted',
        'Harmonic mean of P&R',
        'Area under ROC curve'
    ]
}

performance_df = pd.DataFrame(performance_data)
print("\n" + performance_df.to_string(index=False))

# Save to CSV
performance_df.to_csv(f'{config.RESULTS_DIR}performance_summary_table.csv', index=False)
print(f"\n Performance summary table saved: performance_summary_table.csv")

# Per-class report
class_names = [config.CLASS_NAMES[i] for i in range(config.NUM_CLASSES)]
class_report = classification_report(y_true, y_pred, target_names=class_names, 
                                    digits=4, zero_division=0)
print("\n" + class_report)

# Save report
with open(f'{config.RESULTS_DIR}classification_report.txt', 'w') as f:
    f.write("CNN MODEL - CLASSIFICATION REPORT\n")
    f.write("="*80 + "\n\n")
    f.write(f"Overall Metrics:\n")
    f.write(f"  Accuracy:  {accuracy:.4f}\n")
    f.write(f"  Precision: {precision:.4f}\n")
    f.write(f"  Recall:    {recall:.4f}\n")
    f.write(f"  F1-Score:  {f1:.4f}\n\n")
    f.write(class_report)

# =============================================================================
# PREDICTION VISUALIZATION WITH IMAGES
# =============================================================================

print("\n" + "="*80)
print("VISUAL ANALYSIS: PREDICTIONS")
print("="*80)

def show_predictions(num_correct=5, num_incorrect=5):
    """Show sample predictions with actual images"""
    
    # Find correct and incorrect predictions
    # FIX: Changed y_val to y_true
    correct_idx = np.where(y_pred == y_true)[0]
    incorrect_idx = np.where(y_pred != y_true)[0]
    
    print(f"\n Total Correct: {len(correct_idx)}")
    print(f" Total Incorrect: {len(incorrect_idx)}")
    
    # Sample indices
    correct_samples = np.random.choice(correct_idx, 
                                      min(num_correct, len(correct_idx)), 
                                      replace=False)
    incorrect_samples = np.random.choice(incorrect_idx, 
                                        min(num_incorrect, len(incorrect_idx)), 
                                        replace=False)
    
    # Plot correct predictions
    if len(correct_samples) > 0:
        fig, axes = plt.subplots(1, len(correct_samples), figsize=(3*len(correct_samples), 3))
        if len(correct_samples) == 1:
            axes = [axes]
        
        for idx, sample_idx in enumerate(correct_samples):
            img_file = val_df.iloc[sample_idx]['file']
            img_path = os.path.join(config.VAL_IMAGE_DIR, img_file)
            
            try:
                img = Image.open(img_path)
                img_array = np.array(img)
                
                if len(img_array.shape) == 2:
                    axes[idx].imshow(img_array, cmap='gray')
                else:
                    axes[idx].imshow(img_array)
                
                # FIX: Changed y_val to y_true
                true_label = y_true[sample_idx]
                pred_label = y_pred[sample_idx]
                confidence = y_pred_proba[sample_idx, pred_label] * 100
                
                axes[idx].set_title(f' Correct\nTrue: {true_label}\nPred: {pred_label}\nConf: {confidence:.1f}%',
                                  fontsize=9, color='green', fontweight='bold')
                axes[idx].axis('off')
            except Exception as e:
                axes[idx].text(0.5, 0.5, f'Error loading\nimage', 
                              ha='center', va='center', fontsize=9)
                axes[idx].axis('off')
        
        plt.suptitle('Correct Predictions', fontsize=14, fontweight='bold', color='green')
        plt.tight_layout()
        plt.savefig(f'{config.RESULTS_DIR}correct_predictions.png', dpi=300, bbox_inches='tight')
        plt.show()
    
    # Plot incorrect predictions
    if len(incorrect_samples) > 0:
        fig, axes = plt.subplots(1, len(incorrect_samples), figsize=(3*len(incorrect_samples), 3))
        if len(incorrect_samples) == 1:
            axes = [axes]
        
        for idx, sample_idx in enumerate(incorrect_samples):
            img_file = val_df.iloc[sample_idx]['file']
            img_path = os.path.join(config.VAL_IMAGE_DIR, img_file)
            
            try:
                img = Image.open(img_path)
                img_array = np.array(img)
                
                if len(img_array.shape) == 2:
                    axes[idx].imshow(img_array, cmap='gray')
                else:
                    axes[idx].imshow(img_array)
                
                # FIX: Changed y_val to y_true
                true_label = y_true[sample_idx]
                pred_label = y_pred[sample_idx]
                confidence = y_pred_proba[sample_idx, pred_label] * 100
                
                axes[idx].set_title(f' Incorrect\nTrue: {true_label}\nPred: {pred_label}\nConf: {confidence:.1f}%',
                                  fontsize=9, color='red', fontweight='bold')
                axes[idx].axis('off')
            except Exception as e:
                axes[idx].text(0.5, 0.5, f'Error loading\nimage', 
                              ha='center', va='center', fontsize=9)
                axes[idx].axis('off')
        
        plt.suptitle('Incorrect Predictions (Misclassifications)', fontsize=14, fontweight='bold', color='red')
        plt.tight_layout()
        plt.savefig(f'{config.RESULTS_DIR}incorrect_predictions.png', dpi=300, bbox_inches='tight')
        plt.show()

# Show predictions
show_predictions(num_correct=5, num_incorrect=5)

print("\n Prediction visualizations saved!")

# =============================================================================
# 9. CONFUSION MATRIX
# =============================================================================

print("\n Generating confusion matrix...")

cm = confusion_matrix(y_true, y_pred)

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Raw counts
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=class_names, yticklabels=class_names, cbar_kws={'label': 'Count'})
axes[0].set_xlabel('Predicted Label', fontsize=12, fontweight='bold')
axes[0].set_ylabel('True Label', fontsize=12, fontweight='bold')
axes[0].set_title('Confusion Matrix (Counts)', fontsize=14, fontweight='bold')

# Normalized
cm_norm = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-10)
sns.heatmap(cm_norm, annot=True, fmt='.2%', cmap='RdYlGn', ax=axes[1],
            xticklabels=class_names, yticklabels=class_names, vmin=0, vmax=1)
axes[1].set_xlabel('Predicted Label', fontsize=12, fontweight='bold')
axes[1].set_ylabel('True Label', fontsize=12, fontweight='bold')
axes[1].set_title('Confusion Matrix (Normalized)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{config.RESULTS_DIR}confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# =============================================================================
# 10. ROC CURVES
# =============================================================================

print("\n Generating ROC curves...")

y_true_bin = label_binarize(y_true, classes=range(config.NUM_CLASSES))

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(config.NUM_CLASSES):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

fpr["micro"], tpr["micro"], _ = roc_curve(y_true_bin.ravel(), y_pred_proba.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

plt.figure(figsize=(10, 7))
plt.plot(fpr["micro"], tpr["micro"],
         label=f'Micro-average (AUC = {roc_auc["micro"]:.3f})',
         color='deeppink', linestyle='--', linewidth=3)

colors = plt.cm.tab10(np.linspace(0, 1, config.NUM_CLASSES))
for i, color in zip(range(config.NUM_CLASSES), colors):
    plt.plot(fpr[i], tpr[i], color=color, linewidth=1.5,
             label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', linewidth=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12, fontweight='bold')
plt.ylabel('True Positive Rate', fontsize=12, fontweight='bold')
plt.title('ROC Curves - CNN Model', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=9)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(f'{config.RESULTS_DIR}roc_curves.png', dpi=300, bbox_inches='tight')
plt.show()

print(f" Micro-average AUC: {roc_auc['micro']:.4f}")

# =============================================================================
# 11. TEST SET PREDICTIONS
# =============================================================================

print("\n" + "="*80)
print("SECTION 9: TEST SET PREDICTIONS")
print("="*80)

test_generator = test_datagen.flow_from_dataframe(
    dataframe=test_df,
    directory=config.TEST_IMAGE_DIR,
    x_col='file',
    target_size=(config.IMG_HEIGHT, config.IMG_WIDTH),
    batch_size=config.BATCH_SIZE,
    class_mode=None,
    color_mode=color_mode,
    shuffle=False
)

test_pred_proba = model.predict(test_generator, steps=len(test_generator), verbose=0)
test_pred = np.argmax(test_pred_proba, axis=1)
test_confidence = np.max(test_pred_proba, axis=1)

submission_df = test_df.copy()
submission_df['predicted_label'] = test_pred
submission_df['confidence'] = test_confidence

submission_df.to_csv(f'{config.RESULTS_DIR}test_predictions.csv', index=False)
print(f" Test predictions saved ({len(test_df)} samples)")

# =============================================================================
# 12. FINAL SUMMARY
# =============================================================================

total_time = time.time() - start_time_total

print("\n" + "="*80)
print(" CNN TRAINING COMPLETED SUCCESSFULLY!")
print("="*80)

summary = {
    'Approach': 'Convolutional Neural Network (CNN)',
    'Dataset': {
        'Training Samples': len(train_df),
        'Validation Samples': len(val_df),
        'Test Samples': len(test_df),
        'Classes': config.NUM_CLASSES,
        'Imbalance Ratio': f"{imbalance_ratio:.2f}:1"
    },
    'Preprocessing': {
        'Input Size': f'{input_shape}',
        'Normalization': 'Rescale to [0, 1]',
        'Augmentation': 'Rotation, Shift, Zoom, Flip'
    },
    'Model': {
        'Architecture': '3 Conv Blocks + 2 Dense',
        'Parameters': f'{total_params:,}',
        'Optimizer': f'Adam (lr={config.LEARNING_RATE})',
        'Loss': 'Categorical Crossentropy'
    },
    'Training': {
        'Epochs': len(history.history['loss']),
        'Batch Size': config.BATCH_SIZE,
        'Time': f'{train_time:.2f}s ({train_time/60:.2f} min)',
        'Best Epoch': int(np.argmax(history.history['val_accuracy']) + 1)
    },
    'Performance': {
        'Accuracy': f'{accuracy:.4f}',
        'Precision': f'{precision:.4f}',
        'Recall': f'{recall:.4f}',
        'F1-Score': f'{f1:.4f}',
        'AUC': f'{roc_auc["micro"]:.4f}'
    },
    'Implementation': {
        'Language': 'Python 3',
        'Framework': f'TensorFlow {tf.__version__}',
        'Environment': 'Local/Colab',
        'Random Seed': RANDOM_SEED
    }
}

print(f"\n Assignment Summary:")
print(f"   Total Time: {total_time/60:.2f} minutes")
print(f"   Training Time: {train_time/60:.2f} minutes")
print(f"   Final Accuracy: {accuracy*100:.2f}%")
print(f"   Model Parameters: {total_params:,}")

# Save comprehensive summary
with open(f'{config.RESULTS_DIR}assignment_summary.json', 'w') as f:
    json.dump(summary, f, indent=4)

# Save comparison table
comparison_df = pd.DataFrame({
    'Model': ['CNN'],
    'Accuracy': [f'{accuracy:.4f}'],
    'Precision': [f'{precision:.4f}'],
    'Recall': [f'{recall:.4f}'],
    'F1-Score': [f'{f1:.4f}'],
    'AUC': [f'{roc_auc["micro"]:.4f}'],
    'Training Time': [f'{train_time/60:.1f} min'],
    'Parameters': [f'{total_params:,}'],
    'Notes': ['Deep Learning approach']
})

comparison_df.to_csv(f'{config.RESULTS_DIR}model_comparison.csv', index=False)

print(f"\n All Results Saved to: {config.RESULTS_DIR}")
print("\n Generated Files for Report:")
print("   1. assignment_summary.json - Complete technical details")
print("   2. classification_report.txt - Performance metrics")
print("   3. training_log.csv - Epoch-by-epoch data")
print("   4. model_comparison.csv - For comparing with SVM")
print("   5. test_predictions.csv - Test set results")
print("\n Visualizations for Report:")
print("   1. class_distribution.png - Dataset analysis")
print("   2. training_curves.png - Learning curves (loss vs epochs)")
print("   3. confusion_matrix.png - Model performance")
print("   4. roc_curves.png - ROC with AUC scores")
print("   5. correct_predictions.png - Visual examples of correct predictions")
print("   6. incorrect_predictions.png - Visual examples of misclassifications")

print("\n" + "="*80)
print(" READY FOR ASSIGNMENT SUBMISSION!")
print("="*80)
print(f"\nEnd Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Total Execution Time: {total_time/60:.2f} minutes")
print("\n All sections completed:")
print("   ✓ Dataset Handling")
print("   ✓ Data Preprocessing")
print("   ✓ Model Architecture")
print("   ✓ Training")
print("   ✓ Performance Evaluation")
print("   ✓ Visualizations")
print("   ✓ Test Predictions")
print("\n" + "="*80)
```
